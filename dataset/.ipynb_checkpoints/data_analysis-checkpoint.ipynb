```python
kmean_clustering  = KMeans(n_clusters=14)
```

```python
kmean_clustering.fit(df_users.drop(columns=['user_id']))
```

```python
kmean_clustering.cluster_centers_
```

```python
df_users_labelled = df_users.join(pd.DataFrame(kmean_clustering.labels_, columns=['cluster']))
```
